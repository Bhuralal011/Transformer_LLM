```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# This will now successfully download the model configurations and weights!
model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-2-7b-hf", dtype="auto", device_map="auto")
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-hf")
```

```python
model_inputs = tokenizer(["The secret to baking a good cake is "], return_tensors="pt").to(model.device)
```

```python
model_inputs
```

```python
#For inference, pass the tokenized inputs to generate() to generate text. Decode the token ids back into text with batch_decode().
#This method is provided for backwards compatibility. The decode method now handles batched input natively, so you can use decode directly instead of batch_decode.
```

```python
generated_ids = model.generate(**model_inputs, max_length=30)
tokenizer.batch_decode(generated_ids)[0]
'<s> The secret to baking a good cake is 100% in the preparation. There are so many recipes out there,'
```

```python
# Imports Hugging Face's high-level task runner for easy model text generation.
from transformers import pipeline
# Imports the hardware utility library to automatically configure system resources.
from accelerate import Accelerator

# Scans the system and automatically selects the fastest available hardware (GPU or CPU).
device = Accelerator().device

# Downloads, loads, and moves the 14GB Llama-2 model onto the selected hardware device.
generator = pipeline("text-generation", model="meta-llama/Llama-2-7b-hf", device=device)
```

```python
# from transformers import pipeline

# 1. Define the pipeline task first (it defaults to a standard free model like GPT-2)
generator = pipeline("text-generation")

# 2. Pass your text prompt into the generator variable
result = generator("The secret to baking a good cake is ", max_length=50)

# 3. Print the output
print(result)



# pipeline("The secret to baking a good cake is ", max_length=50)
# [{'generated_text': 'The secret to baking a good cake is 100% in the batter. The secret to a great cake is the icing.\nThis is why we’ve created the best buttercream frosting reci'}]
```

```python
from transformers import pipeline, BitsAndBytesConfig
import torch
```

```python


quant_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16
)

# Changed model name to "Qwen/Qwen3-8B"
generator = pipeline(
    "text-generation",
    model="Qwen/Qwen3-8B",
    model_kwargs={"quantization_config": quant_config}
)

output = generator("The secret to baking a good cake is", max_new_tokens=20)
print(output[0]['generated_text'])






```
